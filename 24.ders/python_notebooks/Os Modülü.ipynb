# Os Modülü

Os modülü işletim sisteminde işlemler gerçekleştirebildiğimiz standard bir Python modülüdür. Bu modül çok kapsamlı bir olduğu için bu derste sadece temel fonksiyonlarını kullanmaya çalışacağız. Burada daha sonra çalışmanız için fonksiyonlarımızın açıklaması bulabilirsiniz.

***Buradaki kullanımları videoda görmeye çalışacağız.***


```python
import os
dir(os)
```

### getcwd() fonksiyonu

Bu fonksiyon, bulunduğumuz dizinin yolunu söyler.



```python
os.getcwd()
```

### chdir() fonksiyonu

Bu fonksiyon , bulunduğumuz dizini değiştirmemizi sağlar.

```python
os.chdir("C:/Users/user/Desktop")
os.getcwd()
```

### listdir() fonksiyonu

Bu fonksiyon, bulunduğumuz dizinde klasörleri ve dosyaları listeler.


```python
os.getcwd()

print(os.listdir())
```

### mkdir() fonksiyonu

Bu fonksiyon , "klasör" oluşturmamızı sağlar.



```python
os.mkdir("Deneme1")
```

### makedirs() fonksiyonu

Bu fonksiyon iç içe klasör oluşturmamızı sağlar.



```python
os.makedirs("Deneme2/Deneme3/Deneme4")
```

### rmdir() fonksiyonu

Bu fonksiyon, klasör silmemizi sağlar.

```python
os.rmdir("Deneme1")
```

### removedirs() fonksiyonu

Bu fonksiyon, iç içe klasörleri silmemizi sağlar.

```python
os.removedirs("Deneme2/Deneme3")
```

### rename() fonksiyonu

Bu fonksiyon, dosya ve klasörlerin ismini değiştirmemizi sağlar.


```python
os.rename("test.txt","test2.txt") # Dosyamızın yeni ismi "test2.txt" oldu.
```

### stat() fonksiyonu 

Bu fonksiyon , bir dosyanın değiştirilme zamanı, büyüklüğü gibi özellikleri veren bir stat objesi döner.

```python
os.stat("test2.txt")
```

### walk() fonksiyonu

Bu fonksiyon belki de os modülünün içindeki en yararlı fonksiyonlardan bir tanesi. Bu fonksiyonumuz içine bir tane dizin (veya klasör) verdiğimizde bunun altındaki tüm dizinleri , dosyaları sıralar ve alt dizinlere gidilecek bir yer kalmayana kadar gider. ***os.walk()*** çıktısını görmek için for döngüsü kullanmaya çalışalım.

```python
for klasör_yolu,klasör_isimleri,dosya_isimleri in os.walk("C:/Users/user/Desktop"):


    print("Şu anki dizin",klasör_yolu)
    print("Klasörler ",klasör_isimleri)
    print("Dosyalar",dosya_isimleri)
    print("**********************************")
```
