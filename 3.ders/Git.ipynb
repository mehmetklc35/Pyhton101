Git, bir versiyon kontrol sistemidir. Yazacağınız programların bir kopyasını alarak daha sonra ihtiyaç halinde aldığınız bu kopyalara (versiyonlara) geri dönebilmek için kullanabilirsiniz. Giti, daha çok yazılım geliştiriciler kullanmaktadır. Ama giti herhangi yazacağınız bir makalede, projede de kullanabilirsiniz. Aldığınız kopyalar sayesinde geriye dönük bir dökümanınız olmuş olur. Git çekirdeği linux'un yaratıcısı Linus Torvalds tarafından yazılmıştır. Açık kaynak bir yazılım ürünüdür, bu sayede ücretsizdir.  Peki git'e neden ihtiyaç duyarız?

Projeleri geliştirirken eski kopyalara da ihtiyaç duyabiliriz. sürekli Copy-Paste yaparsak, karmaşık dosya sistemleri yaratmış oluruz. Git sistemi bu kadar karışık dosyalar yaratmadan projenizi daha profesyonel bir şekilde geliştirmenizi sağlar.
